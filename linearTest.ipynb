```python
import pandas as pd
df = pd.read_csv("C:\\Users\\AANCHAL\\Desktop\\ExcelR\\Datasets\\Toyoto_Corrola.csv")
print(df.head())
```

```python
column = []
for x in df.columns:
    if df[x].dtypes=='object':
        column.append(x)
column
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

for x in column:
    df[x]=le.fit_transform(df[x])


    le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    print('Feature', x)
    print('mapping', le_name_mapping)

```

```python
x = df[['Id', 'Model', 'Age_08_04', 'KM', 'HP', 'Doors', 'Cylinders', 'Gears', 'Weight']]
y = df[['Price']]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
model1.fit(x_train, y_train)
```

```python
print("Beta values :" , model1.coef_)
print("Beta zero value : ", model1.intercept_)
```

```python
df.head()
```

```python
y_pred = model1.predict(x_test)

y_pred
```

```python
x_test['actual values'] = y_test
x_test['predicted values'] = y_pred
x_test
```

```python
from sklearn.metrics import root_mean_squared_error

rmse = root_mean_squared_error(y_test, y_pred)
print(rmse)
```
